# Find runs with high failure rate (>= 70%)

```python
run_rel = exr.stats.groupby(exr.RUN_VARS)[["Request Count","Failure Count"]].sum().reset_index()
run_rel["fr"] = run_rel["Failure Count"] / run_rel["Request Count"]
run_rel["path"] = run_rel.apply(lambda x: f'{x["exp_start"]}/{x["exp_workload"]}/{x["exp_branch"]}/{x["run_iteration"]}', axis=1)

threshold = 0.8
```

```python
# any issues with either WL or branch?
display(run_rel[run_rel["fr"] > threshold].groupby(["exp_branch"])["fr"].count())
display(run_rel[run_rel["fr"] > threshold].groupby(["exp_workload"])["fr"].count())

run_rel[run_rel["fr"] > threshold]["fr"].count() / run_rel["fr"].count()
```

```python
experiment_runs = exr.stats.groupby(["exp_workload","exp_branch"])["run_iteration"].nunique().reset_index()

expexted_runs = pd.DataFrame(list(itertools.product(exr.stats["exp_workload"].unique(), exr.stats["exp_branch"].unique())),columns=["exp_workload","exp_branch"])
experiment_runs = experiment_runs.merge(expexted_runs, how="outer", on=["exp_workload","exp_branch"]).fillna(0)
experiment_runs[experiment_runs["run_iteration"] <= 1]
```
