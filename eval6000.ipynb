# Wattage per Request / Resource Consumption

```python


wdf = exr.pods_energy().merge(
    how="left",
    right=exr.run_stats(),
    on=exr.RUN_VARS,
    validate="one_to_one",
)


wdf['ws_per_rq'] = wdf['wattage_kepler'] / wdf['Success Count']
wdf['mems_per_rq'] = wdf['memory_usage'] / wdf['Success Count']
wdf['cpus_per_rq'] = wdf['cpu_usage'] / wdf['Success Count']

wdf.describe()
```

```python
import matplotlib.pyplot as plt
cols = ['ws_per_rq', 'cpus_per_rq', 'mems_per_rq']
# fig, axs = plt.subplots(ncols=3)

# i = 0
# for i, col in enumerate(cols):
#     baseline = wdf[wdf['exp_branch'] == 'baseline_vanilla_full'][col].mean()
#     display(baseline)
#     ax = sns.barplot(data=wdf, hue='exp_branch', y=col, ax=axs[i]) # , hue="exp_workload")
#     ax.axhline(y=baseline)

wdfm = wdf.melt(id_vars=exr.RUN_VARS, value_vars=cols)
wdfm

g = sns.catplot(
    data=wdfm,
    kind="bar",
    col="variable",
    y="value",
    sharey=False,
    hue="exp_branch",
    palette="rocket",
)

# # baseline magic
for i, col in enumerate(cols):
    baseline = wdf[wdf['exp_branch'] == 'baseline_vanilla_full'][col].mean()
    g.axes[0][i].axhline(baseline, ls="--")


plt.savefig("figs/branch_resources.pdf")

```

```python
hc = exr.stats_history.groupby([*exr.RUN_VARS, "timestamp", "url", "type"]).agg(
    n_rq = ("rq", "count"),
    min_rq = ("rq", "min"),
    max_rq = ("rq", "max") 
).reset_index()
hc[(hc['n_rq'] > 1)]
```

```python
# try to combine history and measurements, e.g. for CPU vs. rq/s

exr.pods['run_seconds'] = exr.pods['run_time'].dt.total_seconds()

run_starts = exr.stats_history.groupby(exr.RUN_VARS)['timestamp'].transform("min")
exr.stats_history['run_seconds'] = exr.stats_history['timestamp'] - run_starts

def round_down_to_nearest(self, n):
    return (self // n) * n

pd.Series.round_down_to_nearest = round_down_to_nearest

exr.stats_history['run5s'] = exr.stats_history['run_time'].round_down_to_nearest(5)
exr.stats_history

runhist = exr.stats_history.groupby([*exr.RUN_VARS, 'run_time']).agg({
    "rq": "sum",
    "user_count": "sum",
})
```

```python
sns.relplot(
    kind="line",
    data=runhist,
    x="run_time",
    y="rq",
    row="exp_branch",
    col="exp_workload",
)
```
