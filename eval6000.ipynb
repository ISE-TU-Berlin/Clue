```python
awesomepods = exr.pods
awesomepods = awesomepods[awesomepods['name'] != 'loadgenerator']
awesomepods = awesomepods[awesomepods['exp_branch'] == "baseline_vanilla_full"]
# awesomepods = awesomepods[awesomepods['exp_workload'] == "exp_scale_pausing"]
# exr.pods['name'].unique()
awesomepods['pod_name'] = awesomepods['name'].apply(lambda x: x[:-2])
# awesomepods = awesomepods[awesomepods['pod_name'] == "teastore-webui"]


podstime = awesomepods.groupby(['exp_workload', 'pod_name', 'run_time']).agg({"wattage_scaph": "mean", "wattage_kepler": "mean", "cpu_usage": "sum", 'name': 'nunique'})

podstime
```

```python
sns.lineplot(data=podstime, x='run_time', y='cpu_usage', hue='exp_workload')
```

```python
# comparing summing up all wattages (assuming they are per second) vs average watt per experiment times number of seconds

node_watts = exr.nodes_energy().reset_index()
wcols = node_watts.columns[node_watts.columns.str.startswith('wattage')]
wcols

kep = node_watts['wattage_kepler'].sum()
sca = node_watts['wattage_scaph'].sum()
tapo = node_watts['wattage'].sum()

ratio = kep/sca

display(kep, sca, tapo, ratio)

sns.barplot(
    data=node_watts.melt(id_vars=exv2.experiment_results.ExperimentResults.RUN_VARS, value_vars=list(wcols)),
    x="exp_workload",
    y="value",
    hue="variable"
)
```
